```python
import pandas as pd
import numpy as np

# --- 1. AMOSTRAGEM ESTRATIFICADA ---
PORCENTAGEM_AMOSTRA = 0.1  # 10%

print("1. Selecionando estações...")
# Garante que a lista de estações para sorteio é única
estacoes_unicas = df[['estado', 'EstacaoCodigo']].drop_duplicates()
estacoes_sorteadas = estacoes_unicas.groupby('estado').sample(frac=PORCENTAGEM_AMOSTRA, random_state=42)

# Filtra os dados
df_amostra = df[df['EstacaoCodigo'].isin(estacoes_sorteadas['EstacaoCodigo'])].copy()

# --- BLINDAGEM 1: Remove duplicatas de MÊS ---
# Se houver duas linhas para o mesmo Ano/Mes na mesma estação, mantemos a primeira
linhas_antes = len(df_amostra)
df_amostra = df_amostra.drop_duplicates(subset=['EstacaoCodigo', 'Ano', 'Mes'])
print(f"Linhas removidas (duplicatas de mês): {linhas_antes - len(df_amostra)}")


# --- 2. PREPARAÇÃO DOS DADOS (MELT) ---
print("2. Organizando e limpando datas...")

colunas_chuva = [f'Chuva{i:02d}' for i in range(1, 32)]

df_melted = df_amostra.melt(
    id_vars=['EstacaoCodigo', 'estado', 'Ano', 'Mes'], 
    value_vars=colunas_chuva,
    var_name='Dia_str',
    value_name='Valor'
)

# Limpeza básica
df_melted = df_melted.dropna(subset=['Valor'])
df_melted['Dia'] = df_melted['Dia_str'].str.replace('Chuva', '').astype(int)

# Cria a Data Real
df_melted['Data_Real'] = pd.to_datetime(
    dict(year=df_melted.Ano, month=df_melted.Mes, day=df_melted.Dia), 
    errors='coerce'
)

# Remove datas inválidas (ex: 30 de fevereiro)
df_melted = df_melted.dropna(subset=['Data_Real'])

# --- BLINDAGEM 2: Remove duplicatas de DIA ---
# Garante exata unicidade do dia para o cálculo de intervalo funcionar
len_antes_dia = len(df_melted)
df_melted = df_melted.drop_duplicates(subset=['EstacaoCodigo', 'Data_Real'])
print(f"Registros diários removidos (duplicatas reais): {len_antes_dia - len(df_melted)}")

# Ordenação obrigatória para o .diff() funcionar
df_melted = df_melted.sort_values(['EstacaoCodigo', 'Data_Real'])


# --- 3. CÁLCULOS POR ANO ---
print("3. Calculando métricas...")

grupos = df_melted.groupby(['EstacaoCodigo', 'Ano'])

# Métrica A: Intervalos (Gaps)
# .diff() calcula a diferença entre a data atual e a anterior
df_melted['gap'] = grupos['Data_Real'].diff().dt.days

# Agregamos estatísticas
resumo_intervalos = grupos['gap'].agg(
    Intervalo_Max='max',
    Intervalo_Medio='mean',
    Intervalo_Min='min'
).fillna(0) # Se só tiver 1 dia, o gap é NaN, trocamos por 0

# Métrica B: Completude
contagem_dias = grupos.size().to_frame(name='dias_com_dados')

# Junta tudo
df_final = contagem_dias.join(resumo_intervalos).reset_index()

# Recupera o estado
df_final = df_final.merge(estacoes_sorteadas, on='EstacaoCodigo', how='left')

# Cálculos finais
def get_dias_ano(ano):
    return 366 if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0) else 365

df_final['total_dias_ano'] = df_final['Ano'].apply(get_dias_ano)
df_final['Completude_%'] = (df_final['dias_com_dados'] / df_final['total_dias_ano']) * 100

# Arredondamentos
df_final['Completude_%'] = df_final['Completude_%'].round(2)
df_final['Intervalo_Medio'] = df_final['Intervalo_Medio'].round(2)

# Organização final
colunas = ['EstacaoCodigo', 'estado', 'Ano', 'Completude_%', 'Intervalo_Max', 'Intervalo_Medio', 'dias_com_dados']
df_final = df_final[colunas]

print("\nAnálise Finalizada sem duplicatas!")
```

```python
import pandas as pd
import numpy as np

# --- CONFIGURAÇÃO ---
PORCENTAGEM_AMOSTRA = 0.1  # 10%
SEEDS_PARA_TESTAR = [42, 1, 100, 2024, 999] # Números aleatórios diferentes para testar
resultados_teste = []

print(f"Iniciando Teste de Estabilidade com {len(SEEDS_PARA_TESTAR)} amostras diferentes...\n")

# Função auxiliar para dias no ano
def get_dias_ano(ano):
    return 366 if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0) else 365

# --- LOOP DE VALIDAÇÃO ---
for seed in SEEDS_PARA_TESTAR:
    
    # 1. AMOSTRAGEM (Mudando o random_state)
    estacoes_unicas = df[['estado', 'EstacaoCodigo']].drop_duplicates()
    
    # AQUI ESTÁ O SEGREDO: Usamos o 'seed' do loop
    estacoes_sorteadas = estacoes_unicas.groupby('estado').sample(frac=PORCENTAGEM_AMOSTRA, random_state=seed)
    
    df_amostra = df[df['EstacaoCodigo'].isin(estacoes_sorteadas['EstacaoCodigo'])].copy()
    
    # Blindagem 1 (Mês)
    df_amostra = df_amostra.drop_duplicates(subset=['EstacaoCodigo', 'Ano', 'Mes'])
    
    # 2. MELT (Prepara os dados)
    colunas_chuva = [f'Chuva{i:02d}' for i in range(1, 32)]
    
    df_melted = df_amostra.melt(
        id_vars=['EstacaoCodigo', 'Ano', 'Mes'], # Removi 'estado' daqui para economizar, recuperamos depois se precisar
        value_vars=colunas_chuva,
        var_name='Dia_str',
        value_name='Valor'
    )
    
    # Limpeza
    df_melted = df_melted.dropna(subset=['Valor'])
    df_melted['Dia'] = df_melted['Dia_str'].str.replace('Chuva', '').astype(int)
    
    # Data Real
    df_melted['Data_Real'] = pd.to_datetime(
        dict(year=df_melted.Ano, month=df_melted.Mes, day=df_melted.Dia), 
        errors='coerce'
    )
    df_melted = df_melted.dropna(subset=['Data_Real'])
    
    # Blindagem 2 (Dia)
    df_melted = df_melted.drop_duplicates(subset=['EstacaoCodigo', 'Data_Real'])
    
    # 3. CÁLCULO DA COMPLETUDE (Simplificado)
    # Agrupa e conta quantos dias existem
    df_resumo = df_melted.groupby(['EstacaoCodigo', 'Ano']).size().to_frame(name='dias_com_dados').reset_index()
    
    # Calcula %
    df_resumo['total_dias_ano'] = df_resumo['Ano'].apply(get_dias_ano)
    df_resumo['Completude_%'] = (df_resumo['dias_com_dados'] / df_resumo['total_dias_ano']) * 100
    
    # 4. RESULTADO FINAL DA RODADA
    media_geral = df_resumo['Completude_%'].mean()
    
    print(f"Seed {seed}: Média Geral = {media_geral:.4f}%")
    
    resultados_teste.append({
        'Random_State': seed,
        'Media_Completude_%': media_geral
    })

# --- ANÁLISE FINAL ---
df_comparacao = pd.DataFrame(resultados_teste)
print("\n" + "="*40)
print("RESUMO DO TESTE DE ESTABILIDADE")
print("="*40)
print(df_comparacao)
print("-" * 40)

amplitude = df_comparacao['Media_Completude_%'].max() - df_comparacao['Media_Completude_%'].min()
print(f"Diferença entre Maior e Menor resultado: {amplitude:.4f} pontos percentuais")

if amplitude < 1.0:
    print("CONCLUSÃO: Sua amostragem é MUITO ESTÁVEL. Pode confiar em rodar apenas uma vez.")
else:
    print("CONCLUSÃO: Há variabilidade. Considere aumentar a porcentagem da amostra.")
```
